```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# we have launch data as string, we need to convert it to int year, data is both in / or -
# delete launch_data with NAN values
mfData = pd.read_csv('mf_data.csv')
# mfData.head()
mfData = mfData.dropna(subset=['Launch_Date'])
mfData['Launch_Date'] = mfData['Launch_Date'].str.replace('/','-')
mfData['Launch_Date'] = mfData['Launch_Date'].str.split('-').str[2]
mfData['Launch_Date'] = mfData['Launch_Date'].astype(int)
```

```python
# drop the rows with ret_1yr as -
mfData = mfData[mfData['ret_1yr'] != '-']
# convert ret_1yr to float
mfData['ret_1yr'] = mfData['ret_1yr'].astype(float)

# mean of ret_1yr
mfData['ret_1yr'].mean()
```

```python
# make graph of ret_1yr vs Launch_Date
# take average of ret_1yr for each Launch_Date
mfData = mfData.groupby('Launch_Date').mean()
mfData = mfData.reset_index()
# mfData.head()
plt.plot(mfData['Launch_Date'], mfData['ret_1yr'])
plt.xlabel('Launch_Date')
plt.ylabel('ret_1yr')
# plt.title('ret_1yr vs Launch_Date')

plt.savefig('ret_1yr_vs_Launch_Date.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# run a regression on ret_1yr vs ret_3yr, ret_5yr
mfData = pd.read_csv('mf_data.csv')
mfData = mfData.dropna(subset=['Launch_Date'])
mfData['Launch_Date'] = mfData['Launch_Date'].str.replace('/','-')
mfData['Launch_Date'] = mfData['Launch_Date'].str.split('-').str[2]
mfData['Launch_Date'] = mfData['Launch_Date'].astype(int)
mfData = mfData[mfData['ret_1yr'] != '-']
mfData['ret_1yr'] = mfData['ret_1yr'].astype(float)
mfData = mfData[mfData['ret_3yr'] != '-']

mfData['ret_3yr'] = mfData['ret_3yr'].astype(float)
mfData = mfData[mfData['ret_5yr'] != '-']
mfData['ret_5yr'] = mfData['ret_5yr'].astype(float)
mfData = mfData[mfData['Expense_Ratio'] != '-']
mfData['Expense_Ratio'] = mfData['Expense_Ratio'].astype(float)
mfData['AUM'] = mfData['AUM'].astype(float)

mfData = mfData.sort_values(by=['ret_1yr'])

# remove outliers with AUM less than 100 Cr
mfData = mfData[mfData['AUM'] > 100]
# remove ret_1yr outliers
mfData = mfData[mfData['ret_5yr'] > -20]
mfData = mfData[mfData['ret_1yr'] < 20]

# run regression using sklearn
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Create polynomial regression object
regr = linear_model.LinearRegression()
model = make_pipeline(PolynomialFeatures(1), regr)
# Train the model using the training sets

# divide into train and test
# ret_1yr = a + b*ret_3yr + c*ret_5yr
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(mfData[['ret_3yr', 'ret_5yr']], mfData['ret_1yr'], test_size=0.2, random_state=0)

model.fit(X_train, y_train)
# Make predictions using the testing set
ret_1yr_pred = model.predict(X_test)
# draw graph
plt.scatter(y_test, ret_1yr_pred)
plt.xlabel('ret_1yr')
plt.ylabel('ret_1yr_pred')
print('Regression Coefficients: \n', regr.coef_)

# add regrssion coefficient to the graph
plt.rcParams["figure.figsize"] = (10,5)
plt.text(-40, -45, 'Regression Coefficients: \n' + str(regr.coef_), fontsize=12)
plt.text(-30, 0, 'AUM > 100 \n', fontsize=12)
# make the graph look better and bigger

# draw y = x line
x = np.linspace(-30, 20, 100)
plt.plot(x,x , ls="--", c=".3")

# circle the outliers, and draw vertical line, from y = x to y = x + 7
for i in range(len(y_test)):
    if abs(y_test.iloc[i] - ret_1yr_pred[i]) > 8:
        plt.plot(y_test.iloc[i], ret_1yr_pred[i], 'ro')
        # plt.text(y_test.iloc[i], ret_1yr_pred[i], mfData['Scheme_Name'].iloc[i], fontsize=8)
        plt.plot([y_test.iloc[i], y_test.iloc[i]], [y_test.iloc[i], ret_1yr_pred[i]], 'r-', lw=1)
        # write difference
        plt.text(y_test.iloc[i], ret_1yr_pred[i] + 1, str(round(abs(y_test.iloc[i] - ret_1yr_pred[i]), 2)), fontsize=8)


# plt.tight_layout()
# save along with text
plt.savefig('ret_1yr_vs_ret_3yr_ret_5yr_labelled.jpg', bbox_inches='tight')
plt.show()
```
